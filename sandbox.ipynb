#### Is there a correlation between the number of issues resolved in a release and the average score of user reviews (after release)?

1. the basic approach

```python
jsons_path_root = Path('DATAR/release_related/all_jsons')
jsons_paths = list(jsons_path_root.iterdir())

issues_solved = []
avg_scores = []

for path in tqdm(jsons_paths):
    # friendly_path = path.stem.replace('-_-', '/')
    with open(path) as file:
        releases = json.load(file)
    for release in releases:
        # release.pop('release_data')
        # release.pop('release_commit_sha')
        # release['num_commits'] = len(release['commits'])
        # release.pop('commits')
        # release['num_created_issues'] = len(release['created_issues'])
        # release.pop('created_issues')
        # release['num_closed_issues'] = len(release['closed_issues'])
        # release.pop('closed_issues')
        # release.pop('pull_requests')
        # release.pop('contributors')
        # release.pop('apk_metrics')
        issues_solved.append(len(release['closed_issues']))
        avg_scores.append(np.mean([review['score'] for review in release['google_play_reviews']]))  # TODO better metric??

print(len(issues_solved), len(avg_scores))
issues_solved, avg_scores = zip(*[(issue, score) for issue, score in zip(issues_solved, avg_scores) if (not np.isnan(score)) and (issue > 0)])
print(len(issues_solved), len(avg_scores))
# print(issues_solved)
# print(avg_scores)

pearsonr(issues_solved, avg_scores)
```

2. correlation with individual scores

```python
jsons_path_root = Path('DATAR/release_related/all_jsons')
jsons_paths = list(jsons_path_root.iterdir())

issues_solved = []
all_scores = []

for path in tqdm(jsons_paths):
    # friendly_path = path.stem.replace('-_-', '/')
    with open(path) as file:
        releases = json.load(file)
    for release in releases:
        issues_solved.append(len(release['closed_issues']))
        all_scores.append([review['score'] for review in release['google_play_reviews']])

print(len(issues_solved), len(all_scores))
issues_solved, all_scores = zip(*[(issue, scores) for issue, scores in zip(issues_solved, all_scores) if scores and (issue > 0)])
print(len(issues_solved), len(all_scores))
# print(issues_solved)
# print(all_scores)

# Expand issues_resolved to match the length of all_scores
expanded_issues_resolved = np.concatenate([[issues_solved[i]] * len(scores) for i, scores in enumerate(all_scores)])
expanded_scores = np.concatenate(all_scores)
print(expanded_issues_resolved)
print(expanded_scores)

pearsonr(expanded_issues_resolved, expanded_scores)
```
